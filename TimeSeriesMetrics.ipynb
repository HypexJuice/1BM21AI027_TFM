Choose a dataset and demonstrate the following different stochastic techniques like Simple exponential smoothing, simple moving average, holt winters smoothing.
Calculate the evaluation metrics:
MAE
MSE
RMSE
MAPE
Identify trend and seasonal patterns for above forecasting techniques for the given dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import pandas as pd

ads = pd.read_csv('ads.csv', index_col=['Time'], parse_dates=['Time'])
df = pd.read_csv('ads.csv')
Time = df['Time']
```

```python
plt.figure(figsize=(15, 7))
plt.plot(ads.Ads)
plt.title('Ads watched (hourly data)')
plt.grid(True)
plt.show()
```

```python
def simpleExponentialSmoothing(alpha,data):
    new = []
    val = data[0]
    new.append(val)
    for i in range(1,len(data)):
        val = alpha*data[i-1]+(1-alpha)*val
        new.append(val)
    return new
```

```python
def simpleMovingAverage(data):
    new = [data[0]]
    for i in range(1,len(data)-1):
        x = (data[i-1]+data[i]+data[i+1])/3
        new.append(x)
    return new
```

```python
def simpleMovingAverage(series, window):
    sma = series.rolling(window=window).mean()
    print(sma)
    return sma
```

```python
def Mae(data,error):
    return sum(abs(data-error))

def Mse(data,error):
    return np.mean((data-error)**2)

def Rmse(data,error):
    return np.sqrt(np.mean((data-error))**2)

def Mape(data,error):
    return sum((abs(data-error)/data)*100)
```

```python
ses=simpleExponentialSmoothing(0.3,df.iloc[:,1])
sma = simpleMovingAverage(ads,3)

mae = Mae(df.iloc[:,1],ses)
mse = Mse(df.iloc[:,1],ses)
rmse = Rmse(df.iloc[:,1],ses)
mape = Mape(df.iloc[:,1],ses)

print(f"Mean Absolute Error: {mae}\nMean Squared Error: {mse}\nRoot Mean Squared Error: {rmse}\nMean Absolute Percentage Error: {mape}%")
plt.figure(figsize=(15, 7))
plt.plot(df.iloc[:,0],df.iloc[:,1])
plt.plot(df.iloc[:,0],ses, label ="Simple Exponential Smoothing")
plt.plot(df.iloc[:,0],sma, "g", label="Simple Moving Average")
plt.title('Ads watched (hourly data)')
plt.legend()
plt.show()

```
